```python
#Add Matplotlib inline magic command
%matplotlib inline

#Dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd

```

```python
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
#Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

#Display dataframe
pyber_data_df.head()
```

```python
# Create the Urban city Dataframe
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df.head()
```

```python
# Create Suburban and Rural Dataframes

suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

rural_cities_df.head()
```

```python
#Get number of rides for urban cities

urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
#Get number of rides for suburban cities

suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count.head()
```

```python
#Get number of rides for rural cities
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count.head()
```

```python
#Get average fare for urban cities

urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
# Get average fare for suburban cities

suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare.head()
```

```python
#Get average fare for rural cities
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare.head()
```

```python
#Get average number of drivers for each urban city
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head()
```

```python
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count.head()
```

```python
#Build a scatter plot for urban cities
plt.scatter(urban_ride_count, urban_avg_fare, s=10*urban_driver_count,
            c="coral", edgecolor="black", linewidth=1,
            alpha=0.8, label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

#Add the legend
plt.legend()
```

```python
#Build a scatter plot for suburban cities
plt.scatter(suburban_ride_count, suburban_avg_fare, s=10*suburban_driver_count,
            c="skyblue", edgecolor="black", linewidth=1,
            alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

#Add the legend
plt.legend()
```

```python
#Build a scatter plot for rural cities
plt.scatter(rural_ride_count, rural_avg_fare, s=10*rural_driver_count,
            c="gold", edgecolor="black", linewidth=1,
            alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

#Add the legend
plt.legend()
```

```python
#Add the scatter charts for each type of city
# Add in code to fix size of plot
plt.subplots(figsize=(10,6))
plt.scatter(urban_ride_count, urban_avg_fare,
            s=10*urban_driver_count,
            c="coral", edgecolor="black", linewidth=1,
            alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count, suburban_avg_fare, 
            s=10*suburban_driver_count,
            c="skyblue", edgecolor="black", linewidth=1,
            alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count, rural_avg_fare, 
            s=10*rural_driver_count,
            c="gold", edgecolor="black", linewidth=1,
            alpha=0.8, label="Rural")

#Incorporate graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

#Add the legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
                 scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

#Incorporate a text label about the circle size.
plt.text(42,35, "Note: Circle size correlates with driver count per city.", fontsize="12")

#Save the figure.
plt.savefig("Analysis/fig1.png")

#show the plot

plt.show()
```
